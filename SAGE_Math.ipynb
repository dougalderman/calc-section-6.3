```python
x = var('x')
```

```python
y = var('y')
```

```python
def f(x, y): return 2*y^2 + x^2 - x^2*y
```

```python
plot3d(f,(x,-5,5),(y,-5,5),color='blue',opacity=0.8)
```

```python
def fx(x, y): return derivative(f(x, y), x)
```

```python
fx(x, y)
```

```python
def fy(x, y): return derivative(f(x, y), y)
```

```python
fy(x, y)
```

```python
def fxx(x, y): return derivative(f(x, y), x, 2)
```

```python
fxx(x, y)
```

```python
def fyy(x, y): return derivative(f(x, y), y, 2)
```

```python
fyy(x, y)
```

```python
def fxy(x, y): return derivative(fx(x, y), y)
```

```python
fxy(x, y)
```

```python
solve([fx(x, y)==0, fy(x, y)==0], x, y)
```

```python
def D(x, y): return fxx(x, y)*fyy(x, y)-(fxy(x, y))^2
```

```python
D(x, y)
```

```python
D(-2, 1)
```

```python
def D1(x, y): return -4*x^2 - 8*y + 8
```

```python
D1(-2, 1)
```

```python
D1(0, 0)
```

```python
fxx(0, 0) # incorrect answer
```

```python
fxx(x, y)
```

```python
def fxx1(x, y): return -2*y + 2
```

```python
fxx1(0, 0) # correct answer
```

```python
D1(2, 1)
```
